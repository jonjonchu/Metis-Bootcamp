# WTWY Street Team Placement
This is the first project for the [Metis Data Science Bootcamp](www.thisismetis.com/).


## Context
WomenTechWomenYes (WTWY) hosts an annual gala during the summer. To drive attendance, their Street team goes to subway stations in NYC to attract people to sign up for the gala.

## Goal
Optimize the placement of Street teams in NYC subway stations, such that they gather the most amount of signups.
